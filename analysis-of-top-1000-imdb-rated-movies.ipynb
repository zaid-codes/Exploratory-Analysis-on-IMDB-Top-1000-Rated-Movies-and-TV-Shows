# **Exploratory Data Aanlysis on IMDB Dataset of top 1000 movies and tv shows**

# > **Introduction**

This analysis uncovers the insights of the dataset top 1000 movies and TV shows on IMDb. By exploring the data, cleaning it for accuracy, and uncovering hidden insights, to find the trends, patterns, and intriguing relationships.

# > **Data Integrity And Cleaning:**

* Missing Values: Checked for missing values in all columns. Addressed missing data through imputation or removal as appropriate.
* Duplicate Entries: Checked for and removed any duplicate entries in the dataset.
* Outliers: Investigated potential outliers in key variables, addressing them based on the nature of the data.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv(r"/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv")

movies
```

```python
movies.sort_index()
```

```python
movies.describe()
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.isna()
```

Remove the columns having null values or blank cells

```python
movies.dropna(subset = ['Certificate', 'Meta_score', 'Gross'],inplace = True)
movies.dropna(inplace = True)

movies.isnull().sum()
```

# > **Descriptive Statistics**

Changing the type from float to int of column Gross

```python
movies['Gross'] = movies['Gross'].str.replace(',', '')
movies['Gross'] = movies['Gross'].astype('float64')

movies['Gross'] = movies['Gross'].astype(int)
movies['Gross'].dtype
```

```python
movies.describe()
```

Correlation between IMDB_Rating, Meta Score and Gross

```python
selected_columns = ['IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Gross']
selected_data = movies[selected_columns]

# Calculate the correlation matrix
correlation_matrix = selected_data.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
correlation = movies["IMDB_Rating"].corr(movies["No_of_Votes"])
print("Correlation between IMDb rating and number of votes:", correlation)
```

```python
# Average IMDb Rating
average_rating_imdb = movies['IMDB_Rating'].mean()
average_rating_meta = movies['Meta_score'].mean()

# Most Common Genres
most_common_genres = movies['Genre'].value_counts().idxmax()

# Distribution of Box Office Earnings
box_office_distribution = movies['Gross'].describe()

# Display Descriptive Statistics
print(f"Average IMDb Rating: {average_rating_imdb:.2f}")
print(f"Average Meta Score Rating: {average_rating_meta:.2f}")
print(f"Most Common Genres: {most_common_genres}")
print("\nDistribution of Box Office Earnings:")
print(box_office_distribution)
```

```python
hidden_gems = movies[movies["No_of_Votes"] < 100000]  # Adjust threshold as needed
hidden_gems = hidden_gems.sort_values("IMDB_Rating", ascending=False).head(10)
print("Potential Hidden Gems:")
print(hidden_gems[["Series_Title", "IMDB_Rating", "No_of_Votes"]])
```

```python
hidden_gems = movies[movies["No_of_Votes"] < 100000]  # Adjust threshold as needed
hidden_gems = hidden_gems.sort_values("Meta_score", ascending=False).head(10)
print("Potential Hidden Gems:")
print(hidden_gems[["Series_Title", "Meta_score", "No_of_Votes"]])
```

# > **Visualization**

```python
top_movies_imdb = movies.sort_values(by='IMDB_Rating', ascending=False).head(5)

# Sort the DataFrame by Meta score in descending order
top_movies_meta = movies.sort_values(by='Meta_score', ascending=False).head(5)

# Plotting
plt.figure(figsize=(12, 6))

# Plotting IMDb Rating
plt.subplot(1, 2, 1)
sns.barplot(x='IMDB_Rating', y='Series_Title', data=top_movies_imdb, palette='Blues_r')
plt.title('Top 5 Movies Based on IMDb Rating')

# Plotting Meta score
plt.subplot(1, 2, 2)
sns.barplot(x='Meta_score', y='Series_Title', data=top_movies_meta, palette='Reds_r')
plt.title('Top 5 Movies Based on Meta Score')

plt.tight_layout()
plt.show()
```

```python
top_voted_movies = movies.sort_values(by='No_of_Votes', ascending=False).head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x='No_of_Votes', y='Series_Title', data=top_voted_movies, palette='viridis')
plt.title('Top 5 Voted Movies')
plt.xlabel('Number of Votes')
plt.ylabel('Movie Title')
plt.show()
```

```python
top_rated_movies = movies.sort_values(by='IMDB_Rating', ascending=False).head(5)

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='IMDB_Rating', y='Series_Title', data=top_voted_movies, palette='Blues_r')
plt.title('Top 5 Voted Movies based on IMDb Rating')
plt.xlabel('IMDb Rating')
plt.ylabel('Movie Title')
plt.show()
```

```python
top_meta_movies = movies.sort_values(by='Meta_score', ascending=False).head(5)

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='Meta_score', y='Series_Title', data=top_voted_movies, palette='Reds_r')
plt.title('Top 5 Voted Movies based on Meta Score')
plt.xlabel('Meta Score')
plt.ylabel('Movie Title')
plt.show()
```

```python
top_movies_gross = movies.groupby('Series_Title')['Gross'].sum().nlargest(5)

# Create a bar plot for the top 5 directors based on gross earnings
plt.figure(figsize=(25, 10))
sns.barplot(x=top_movies_gross.index, y=top_movies_gross.values, order=top_movies_gross.index)
plt.title('Top 5 Movie with the Highest Total Gross Earnings',)
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 
plt.xlabel('Movie')
plt.ylabel('Total Gross Earnings (in billions)')
plt.xticks(rotation=45, ha='right') # Adjust rotation for better readability
plt.show()
```

Actors and Directors

```python
top_star1 = movies['Star1'].value_counts().nlargest(5)

# Find the top 5 actors (Star 2) with the most appearances
top_star2 = movies['Star2'].value_counts().nlargest(5)

# Combine the top actors from Star 1 and Star 2
top_actors_combined = pd.concat([top_star1, top_star2], axis=1, keys=['Star 1', 'Star 2'])

# Create a bar plot for the top 5 actors based on appearances
plt.figure(figsize=(12, 6))
top_actors_combined.plot(kind='bar', stacked=True)
plt.title('Top 5 Actors with the Most Appearances (Star 1 and Star 2)')
plt.xlabel('Actor')
plt.ylabel('Number of Appearances')
plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
plt.legend(title='Star Position')
plt.show()
```

```python
star1_ratings = movies[['Star1', 'IMDB_Rating']].rename(columns={'Star1': 'Actor', 'IMDB_Rating': 'Rating'})

# Find the top 5 actors with the highest average IMDb ratings
top_actors_ratings = star1_ratings.groupby('Actor')['Rating'].mean().nlargest(5)

# Create a bar plot for the top 5 actors based on average IMDb ratings
plt.figure(figsize=(12, 6))
sns.barplot(x=top_actors_ratings.index, y=top_actors_ratings.values)
plt.title('Top 5 Actors with the Highest Average IMDb Ratings')
plt.xlabel('Actor')
plt.ylabel('Average IMDb Rating')
plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
plt.show()

print("Top 5 Actors with the Highest Average IMDb Ratings:")
print(top_actors_ratings)
```

```python
star3_ratings = movies[['Star1', 'Meta_score']].rename(columns={'Star1': 'Actor', 'Meta_score': 'Rating'})

# Find the top 5 actors with the highest average IMDb ratings
top_actors_ratings1 = star3_ratings.groupby('Actor')['Rating'].mean().nlargest(5)

# Create a bar plot for the top 5 actors based on average IMDb ratings
plt.figure(figsize=(12, 6))
sns.barplot(x=top_actors_ratings1.index, y=top_actors_ratings1.values)
plt.title('Top 5 Actors with the Highest Average Meta Score Ratings ')
plt.xlabel('Actor')
plt.ylabel('Average IMDb Rating')
plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
plt.show()

print("Top 5 Actors with the Highest Average Meta Score Ratings (Star 1 and Star 2):")
print(top_actors_ratings1)
```

```python
top_star1 = movies.groupby('Star1')['No_of_Votes'].sum().nlargest(5)

# Create a bar plot for the top 5 actors based on the total number of votes in movies they appeared in
plt.figure(figsize=(12, 6))
sns.barplot(x=top_star1.index, y=top_star1.values)
plt.title('Top 5 Actors with the Highest Total Number of Votes')
plt.xlabel('Actor (Star 1)')
plt.ylabel('Total Number of Votes')
plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
plt.show()
```

```python
top_star2 = movies.groupby('Star1')['Gross'].sum().nlargest(5)

# Create a bar plot for the top 5 actors based on the total number of votes in movies they appeared in
plt.figure(figsize=(12, 6))
sns.barplot(x=top_star2.index, y=top_star2.values)
plt.title('Top 5 Actors  with the Highest Gross Earnings')
plt.xlabel('Actor (Star 1)')
plt.ylabel('Total Earnings')
plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
plt.show()
```

```python
top_movies_gross = movies.groupby('Director')['Gross'].sum().nlargest(5)

# Create a bar plot for the top 5 directors based on gross earnings
plt.figure(figsize=(25, 10))
sns.barplot(x=top_movies_gross.index, y=top_movies_gross.values, order=top_movies_gross.index)
plt.title('Top 5 Movie Directors with the Highest Total Gross Earnings',)
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 
plt.xlabel('Movie')
plt.ylabel('Total Gross Earnings (in billions)')
plt.xticks(rotation=45, ha='right') # Adjust rotation for better readability
plt.show()
```

```python
from collections import Counter
genre=[]
for x in movies['Genre']:
    for y in x.split(','):
        genre.append(y.strip().lower())

count=Counter(genre)
count=count.most_common()[:5]
x,y=map(list,zip(*count))

fig,axs=plt.subplots(figsize=(20,5))
g=sns.barplot(y=y,x=x)
g.set_ylabel("Genres", weight = "bold")
g.set_title("Top Ten Genres", weight = "bold")
plt.show()
```
